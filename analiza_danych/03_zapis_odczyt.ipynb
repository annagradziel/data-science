```
import numpy as np
import pandas as pd
```

```
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('FB')
df.head()
```

# Zapis/Odczyt danych z pliku CSV

```
df.to_csv('fb.csv')
```

```
df_nov = df[(df.index.month == 11) & (df.index.year == 2019)]
df_nov
```

```
df_nov.to_csv('fb_nov.csv')
```

Wczytywanie danych

```
new_df = pd.read_csv('fb_nov.csv', index_col=0)
new_df
```

Zapisywanie do Excel

```
df_nov.to_excel('fb_nov.xlsx')
```

Wczytywanie danych z excel

```
new_df = pd.read_excel('fb_nov.xlsx', index_col=0)
new_df
```

Ponieważ po wczytaniu danych z Excel tworzy sie nowy index, użyjemy parametru index_col=0 aby usunąć "sztuczny" indeks
